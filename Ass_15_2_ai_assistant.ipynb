### **Task Descriptor-1: Setup FastAPI Backend**

#### **1. Prompt for AI Assistant**

"Generate a basic FastAPI application that includes installing `fastapi` and `uvicorn`, and sets up a root (`/`) GET endpoint that returns `{"message": "Welcome to AI-powered FastAPI"}`."

````python
import subprocess
from fastapi import FastAPI
import nest_asyncio
import uvicorn

# Install FastAPI and Uvicorn
print("Installing FastAPI and Uvicorn...")
process = subprocess.run(["pip", "install", "fastapi", "uvicorn"], capture_output=True, text=True)
print(process.stdout)
if process.stderr:
    print(process.stderr)
print("Installation complete.\n")

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "Welcome to AI-powered FastAPI"}

# For demonstration in Colab, run via nest_asyncio and uvicorn
nest_asyncio.apply()

class Config(uvicorn.Config):
    def __init__(self, app, host="0.0.0.0", port=8000):
        super().__init__(app, host=host, port=port)

config = Config(app)
server = uvicorn.Server(config)
await server.serve() # This line will now execute to start the server
````

#### **1. Expected Output**

```

http://127.0.0.1:8000/

Response:
{"message": "Welcome to AI-powered FastAPI"}
```

### **Task Descriptor-2: API – Update and Delete Items**

#### **1. Prompt for AI Assistant**

"Extend the FastAPI application to include endpoints for updating and deleting items. The `/items` list should continue to be used as temporary storage. Implement a `PUT` endpoint at `/items/{item_id}` that takes an `item_id` (integer) and an updated `item` (dictionary), replacing the item at the specified index. Also, implement a `DELETE` endpoint at `/items/{item_id}` that removes the item at the given `item_id` from the list."

````python
from fastapi import FastAPI, HTTPException
import nest_asyncio
import uvicorn

app = FastAPI()

# Using a Python list as temporary storage for items
# For simplicity, we'll assume item_id corresponds to the list index.
items = []

@app.get("/items")
def get_items():
    return items

@app.post("/items")
def add_item(item: dict):
    items.append(item)
    print(f"Item added: {item}")
    return {"message": "Item created", "item": item}

@app.put("/items/{item_id}")
def update_item(item_id: int, item: dict):
    if 0 <= item_id < len(items):
        items[item_id] = item
        print(f"Item at index {item_id} updated to: {item}")
        return {"message": f"Item {item_id} updated", "item": item}
    raise HTTPException(status_code=404, detail="Item not found")

@app.delete("/items/{item_id}")
def delete_item(item_id: int):
    if 0 <= item_id < len(items):
        deleted_item = items.pop(item_id)
        print(f"Item at index {item_id} deleted: {deleted_item}")
        return {"message": f"Item {item_id} deleted", "item": deleted_item}
    raise HTTPException(status_code=404, detail="Item not found")

# To run this FastAPI application in a Colab environment:
nest_asyncio.apply()

class Config(uvicorn.Config):
    def __init__(self, app, host="0.0.0.0", port=8000):
        super().__init__(app, host=host, port=port)

config = Config(app)
server = uvicorn.Server(config)
await server.serve() # Uncommented to run the server
````

### **Task Descriptor-3: Update Item by Index**

#### **1. Prompt for AI Assistant**

"Generate a FastAPI endpoint to update an existing item in a Python list called `items` using its integer index. The endpoint should be a `PUT` request at `/items/{index}`. If the `index` is out of bounds, return an error message. Otherwise, update the item at that `index` with the provided dictionary `item` and return a success message with the updated item."

````python
from fastapi import FastAPI
import nest_asyncio
import uvicorn

app = FastAPI()

# Using a Python list as temporary storage for items
# For simplicity, we'll assume item_id corresponds to the list index.
items = []

# Adding a dummy item for demonstration purposes
items.append({"name": "Existing Item", "price": 100})

@app.put("/items/{index}")
def update_item(index: int, item: dict):
    if index >= len(items) or index < 0:
        return {"error": "Item not found"}
    items[index] = item
    return {"message": "Item updated", "item": item}

# To run this FastAPI application in a Colab environment:
nest_asyncio.apply()

class Config(uvicorn.Config):
    def __init__(self, app, host="0.0.0.0", port=8000):
        super().__init__(app, host=host, port=port)

config = Config(app)
server = uvicorn.Server(config)
await server.serve() # Uncomment and execute this to run the server
````

### **Task Descriptor-6: Add Auto-Generated Documentation (FastAPI)**

#### **1. Prompt for AI Assistant**

"Generate a FastAPI application that includes two endpoints: a GET endpoint for `/items` to retrieve all items and a POST endpoint for `/items` to add a new item. Ensure both endpoints have clear docstrings that describe their functionality, expected input, and output. Explain how FastAPI leverages these docstrings to automatically generate interactive API documentation."

````python
from fastapi import FastAPI
import nest_asyncio
import uvicorn

app = FastAPI()

# Using a Python list as temporary storage for items
items = []

@app.get(
    "/items",
    summary="Retrieve all items",
    description="Returns a list of all current items in the store."
)
def get_items():
    """
    Retrieve all items.

    This endpoint fetches and returns the entire list of items.
    """
    return items

@app.post(
    "/items",
    summary="Add a new item",
    description="Adds a new item to the store."
)
def add_item(item: dict):
    """
    Add a new item to the list.

    - **item**: A dictionary representing the item to be added (e.g., `{"name": "Laptop", "price": 1200}`).

    Returns a success message with the created item.
    """
    items.append(item)
    return {"message": "Item created", "item": item}

# To run this FastAPI application in a Colab environment:
nest_asyncio.apply()

class Config(uvicorn.Config):
    def __init__(self, app, host="0.0.0.0", port=8000):
        super().__init__(app, host=host, port=port)

config = Config(app)
server = uvicorn.Server(config)
await server.serve() # Uncomment and execute this to run the server
````

#### **3. Expected Output**

*   **Clear documentation for all endpoints:** Each endpoint (`/items` GET and POST) will have its `summary` and `description` displayed.
*   **Endpoint methods:** Clearly indicating `GET` and `POST`.
*   **Input parameters:** For the POST endpoint, it will show the expected `item` (dictionary) as request body.
*   **Sample payloads and responses:** FastAPI automatically infers these from your type hints and docstrings, making it easy for users to understand how to interact with your API.
