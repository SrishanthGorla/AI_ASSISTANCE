```python
# Q1 write a code to check if a number is pallindrome or not

def is_palindrome(n):
    num_str = str(n)
    reversed_str = num_str[::-1]
    return num_str == reversed_str

# Test the function
# enter a number
number = int(input("Enter a number: "))
# conditional statement to check if the number is palindrome
if is_palindrome(number):
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")

#Q2  input: 5 output: 120
# write a code to find  input:5 output:120

def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
# Check the function
number = int(input("Enter a number: "))
result = factorial(number)
print(f"The factorial of {number} is {result}.")

# Q3 WRITE A PYTHON CODE TO FIND •	Input: 153 → Output: Armstrong Number
#Input: 370 → Output: Armstrong Number
#Input: 123 → Output: Not an Armstrong Number

def is_armstrong(n):
    num_str = str(n)
    num_digits = len(num_str)
    total = sum(int(digit) ** num_digits for digit in num_str)
    return total == n

# Test the function
number = int(input("Enter a number: "))
if is_armstrong(number):
    print(f"{number} is an Armstrong Number.")
else:
    print(f"{number} is not an Armstrong Number.")

# Q4 WRITE A CODE TO CHECK WEATHER THE NUMBER IS PERFECT NUMBER OR NOT

def is_perfect_number(n):
    if n < 1:
        return False
    divisors_sum = sum(i for i in range(1, n) if n % i == 0)
    return divisors_sum == n
# Test the function
number = int(input("Enter a number: "))
if is_perfect_number(number):
    print(f"{number} is a Perfect Number.")
else:
    print(f"{number} is not a Perfect Number.")

# Q5 Check even or odd
def is_even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"
# give user input
number = int(input("Enter a number: "))
result = is_even_or_odd(number)
print(f"{number} is {result}.")
# Q6 code to generate prime,composite or neither
#conditional statement to check prime,composite or neither
def check_prime_composite(n):
    if n <= 1:
        return "Neither prime nor composite"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return "Composite"
    return "Prime"
#check the function
number = int(input("Enter a number: "))
result = check_prime_composite(number)
print(f"{number} is {result}.")
```
