Task 1 – Weather API Integration
Connect to a Weather API to fetch current weather details for a given city.

Instructions:

• Prompt AI to generate Python code using the requests library.

• Handle errors: invalid city name, API key errors, and network
failures.

• Display temperature, humidity, and weather description in a
readable format

Prompt:

Generate Python code to fetch current weather data for a given city using a Weather API (like OpenWeatherMap).
The code should:
- Use the requests library
- Handle errors such as invalid city name, API key errors, and network failures
- Display temperature, humidity, and weather description clearly
- Take city name as user input


```python
import requests

# 🔑 Paste your API key here (no spaces)
# You can get one by signing up on OpenWeatherMap: https://openweathermap.org/api
API_KEY = "PASTE_YOUR_REAL_API_KEY_HERE"

def get_weather(city):
    try:
        # ✅ Use HTTPS (important)
        url = "https://api.openweathermap.org/data/2.5/weather"

        params = {
            "q": city,
            "appid": API_KEY,
            "units": "metric"  # You can change to 'imperial' for Fahrenheit
        }

        response = requests.get(url, params=params, timeout=10)

        if response.status_code == 200:
            data = response.json()

            print("\n✅ Weather Details")
            print("----------------------------")
            print(f"City: {data['name']}")
            print(f"Temperature: {data['main']['temp']} °C")
            print(f"Humidity: {data['main']['humidity']} %")
            print(f"Description: {data['weather'][0]['description'].capitalize()}")

        elif response.status_code == 401:
            print("❌ ERROR: Invalid API key OR key not activated yet.")
            print("👉 Double-check your API key and ensure it's active. It might take a few minutes after generation.")

        elif response.status_code == 404:
            print("❌ ERROR: City not found.")

        else:
            print(f"❌ ERROR: {response.status_code} - {response.reason}")
            print(response.text)

    except requests.exceptions.ConnectionError:
        print("❌ Network error. Please check your internet connection.")

    except requests.exceptions.Timeout:
        print("❌ Request timed out. The server did not respond in time.")

    except requests.exceptions.RequestException as e:
        print(f"❌ An unexpected request error occurred: {e}")

    except Exception as e:
        print(f"❌ An unexpected error occurred: {e}")

# Hardcoded city name - change this to your desired city
city = "guntur"
get_weather(city)
```

Task 2 – Currency Exchange Rates

Integrate a Currency Exchange API to convert INR to USD, EUR, and GBP.
Instructions:

• Use AI to generate API request code.

• Validate user input for amount and currency codes.

• Implement error handling for invalid responses and API
downtime.

• Show conversion results clearly in tabular format.

Prompt:

Generate a Python program to convert INR to USD, EUR, and GBP using a Currency Exchange API.

Requirements:

- Use requests library to fetch exchange rates from a free API (like exchangerate-api or open.er-api).

- Take user input for amount in INR.

- Validate input (amount must be positive number).

- Convert INR to USD, EUR, and GBP.

- Handle errors:

  - Invalid input

  - API failure or downtime

  - Invalid API response

- Display results in a clean tabular format.

- Use proper exception handling.

- Make the code beginner-friendly with comments.

```python
import requests

def get_exchange_rates():
    try:
        url = "https://open.er-api.com/v6/latest/INR"
        response = requests.get(url, timeout=5)

        # Check HTTP status
        if response.status_code != 200:
            print("Error: Unable to fetch data from API.")
            return None

        data = response.json()

        # Validate API response
        if data["result"] != "success":
            print("Error: Invalid API response.")
            return None

        rates = data["rates"]

        return {
            "USD": rates.get("USD"),
            "EUR": rates.get("EUR"),
            "GBP": rates.get("GBP")
        }

    except requests.exceptions.Timeout:
        print("Error: API request timed out.")
    except requests.exceptions.ConnectionError:
        print("Error: Network problem.")
    except Exception as e:
        print("Unexpected Error:", e)

    return None


def convert_currency(amount, rates):
    return {
        "USD": amount * rates["USD"],
        "EUR": amount * rates["EUR"],
        "GBP": amount * rates["GBP"]
    }


def display_table(amount, results):
    print("\n====== Currency Conversion ======")
    print(f"{'Currency':<10} {'Amount':>10}")
    print("-" * 25)
    print(f"{'USD':<10} {results['USD']:.2f}")
    print(f"{'EUR':<10} {results['EUR']:.2f}")
    print(f"{'GBP':<10} {results['GBP']:.2f}")
    print("=" * 25)


def main():
    try:
        amount = float(input("Enter amount in INR: "))

        # Input validation
        if amount <= 0:
            print("Error: Amount must be greater than 0.")
            return

    except ValueError:
        print("Error: Please enter a valid number.")
        return

    rates = get_exchange_rates()

    if rates:
        results = convert_currency(amount, rates)
        display_table(amount, results)
    else:
        print("Currency conversion failed due to API error.")


if __name__ == "__main__":
    main()
```

Task 3 – GitHub Repository Info Fetcher
Connect to the GitHub API to fetch details of a repository (e.g., stars,forks, issues).

Instructions:

• Prompt AI to write a Python function to send GET requests to
GitHub API.

• Handle rate limit errors, 404 (repo not found), and invalid input.

• Display repository name, description, stars, forks, and open
issues

Prompt:

Generate a Python program to fetch GitHub repository details using the GitHub REST API.

Requirements:

- Use the requests library.

- Take user input for repository in the format: owner/repo (e.g.,

torvalds/linux).

- Send a GET request to GitHub API: https://api.github.com/repos/

{owner}/{repo}

- Display:

  - Repository Name

  - Description

  - Stars

  - Forks

  - Open Issues

- Handle errors:

  - Invalid input format

  - 404 (repository not found)

  - Rate limit exceeded (status code 403)

  - Network errors

- Display output in a clean readable format.

- Add comments for beginner understanding.

```python
import requests

def fetch_github_repo(repo_name):
    try:
        # Validate input format
        if "/" not in repo_name:
            print("Error: Invalid format. Use 'owner/repo' (e.g., microsoft/Windows-Calculator).")
            return

        url = f"https://api.github.com/repos/{repo_name}"
        response = requests.get(url, timeout=5)

        # Handle status codes
        if response.status_code == 200:
            data = response.json()

            repo_details = {
                "Name": data.get("name"),
                "Description": data.get("description"),
                "Stars": data.get("stargazers_count"),
                "Forks": data.get("forks_count"),
                "Open Issues": data.get("open_issues_count")
            }

            return repo_details

        elif response.status_code == 404:
            print("Error: Repository not found.")
        elif response.status_code == 403:
            print("Error: API rate limit exceeded. Try again later.")
        else:
            print(f"Error: Unexpected status code {response.status_code}")

    except requests.exceptions.Timeout:
        print("Error: Request timed out.")
    except requests.exceptions.ConnectionError:
        print("Error: Network connection issue.")
    except Exception as e:
        print("Unexpected Error:", e)

    return None


def display_repo_info(info):
    if info:
        print("\n====== GitHub Repository Details ======")
        print(f"Name          : {info['Name']}")
        print(f"Description   : {info['Description']}")
        print(f"Stars         : {info['Stars']}")
        print(f"Forks         : {info['Forks']}")
        print(f"Open Issues   : {info['Open Issues']}")
        print("=======================================")


def main():
    repo_name = input("Enter repository (owner/repo): ").strip()

    repo_info = fetch_github_repo(repo_name)

    if repo_info:
        display_repo_info(repo_info)


if __name__ == "__main__":
    main()
```

Task 4 – Real-Time Application: News Headlines Aggregator
Scenario: Build a News Aggregator using a free News API.

Requirements:

• Fetch top 5 headlines for a given category (sports, technology,
health).

• Handle errors like invalid category, missing API key, and request
failures.

• Display headlines in a user-friendly numbered list format.

• Include a retry mechanism if the first request fails.

Prompt:

Use the requests library to send API requests

Handle errors such as:

Invalid category input

Missing or invalid API key

Network/request failures

Display headlines in a clean numbered list format

Implement a retry mechanism (retry up to 3 times if request fails)

Use proper exception handling and user-friendly messages

```python
import requests
import time

# -------------------------------
# CONFIGURATION
# -------------------------------
API_KEY = "https://newsapi.org/account"   # Replace with your NewsAPI key
BASE_URL = "https://newsapi.org/v2/top-headlines"

VALID_CATEGORIES = ["sports", "technology", "health"]

# -------------------------------
# FUNCTION TO FETCH NEWS
# -------------------------------
def get_news(category, retries=3):
    if category.lower() not in VALID_CATEGORIES:
        print("❌ Error: Invalid category. Choose from sports, technology, health.")
        return

    params = {
        "apiKey": API_KEY,
        "category": category.lower(),
        "country": "in",   # India news
        "pageSize": 5
    }

    attempt = 0

    while attempt < retries:
        try:
            response = requests.get(BASE_URL, params=params, timeout=5)

            # Handle API errors
            if response.status_code == 401:
                print("❌ Error: Invalid or missing API key.")
                return
            elif response.status_code != 200:
                print(f"⚠️ Request failed (Status Code: {response.status_code}). Retrying...")
                attempt += 1
                time.sleep(2)
                continue

            data = response.json()

            if data["status"] != "ok":
                print("❌ Error fetching news:", data.get("message", "Unknown error"))
                return

            articles = data.get("articles", [])

            if not articles:
                print("⚠️ No news found for this category.")
                return

            # Display headlines
            print(f"\n📰 Top 5 {category.capitalize()} Headlines:\n")
            for i, article in enumerate(articles, start=1):
                print(f"{i}. {article['title']}")

            return

        except requests.exceptions.RequestException:
            print("⚠️ Network error. Retrying...")
            attempt += 1
            time.sleep(2)

    print("❌ Failed after multiple attempts. Please try again later.")

# -------------------------------
# MAIN PROGRAM
# -------------------------------
if __name__ == "__main__":
    category = input("Enter category (sports/technology/health): ")
    get_news(category)
```
